```python
import pandas as pd
from textblob import TextBlob

# Assuming you have a DataFrame 'df' with a 'text' column
# Replace 'your_file.xlsx' and 'text' with your actual file name and text column name
df = pd.read_excel('dataset.xlsx')

# Apply sentiment analysis to the 'text' column
df['sentiment'] = df['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Categorize sentiment
def categorize_sentiment(score):
    if score > 0.7:
        return 'Strong Positive'
    elif 0.3 <= score <= 0.7:
        return 'Positive'
    elif -0.3 <= score < 0.3:
        return 'Neutral'
    elif -0.7 <= score < -0.3:
        return 'Negative'
    else:
        return 'Strong Negative'

# Create a new column for sentiment category
df['sentiment_category'] = df['sentiment'].apply(categorize_sentiment)

# Display the results
print(df[['text', 'sentiment_category']])
```

```python
import matplotlib.pyplot as plt

# Visualize the distribution of sentiments
plt.hist(df['sentiment'], bins=30, edgecolor='black')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Load your dataset
# Replace 'your_file.xlsx' with your actual file name
df = pd.read_excel('output_sentiment_analysis.xlsx')

# Split the dataset into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Display the size of the training and testing sets
print(f"Training set size: {len(train_df)}")
print(f"Testing set size: {len(test_df)}")
```

```python
# Initialize the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased')

# Create a pipeline with DistilBERT and a logistic regression classifier
text_clf = Pipeline([
    ('tfidf', tokenizer),
    ('clf', LogisticRegression())
])

# Train the model
text_clf.fit(train_df['text'], train_df['label'])
```
